```python
import pandas as pd
```

```python
authors_ = dict(author_id = [1,2,3], author_name = ['Тургенев', 'Чехов', 'Островский'])
```

```python
authors = pd.DataFrame(authors_)
authors
```

```python
book_ = dict(author_id = [1, 1, 1, 2, 2, 3, 3], 
             book_title = ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 
                         'Толстый и тонкий', 'Дама с собачкой', 'Гроза',
                         'Таланты и поклонники'],
             price = [450, 300, 350, 500, 450, 370, 290])
```

```python
book = pd.DataFrame(book_)
book
```

```python
authors_price = pd.merge(book, authors, on="author_id", how="inner")
authors_price
```

```python
top5 = authors_price.nlargest(5, "price")
top5
```
